```python
def make_star(n):
    result = []

    if n == 1:
        return ["*"]

    n //= 3
    S = make_star(n)

    for i in S:
        result.append(i*3)

    for i in S:
        result.append(i+" "*len(S)+i)

    for i in S:
        result.append(i*3)

    return result


stars = make_star(int(input()))
for i in stars:
    print(i)
```

```python
# 10815 숫자카드 try 1

n = int(input())
nlist = sorted(list(map(int,input().split())))
m = int(input())
mlist = list(map(int,input().split()))

mid = n//2

for card in mlist :

    mid = n//2
    target_num = nlist[mid]

    if target_num > card :
        while target_num > card and mid != 0:
            mid //= 2
            target_num = nlist[mid]
        
        if card in nlist[mid:mid*2+1]:
            print(1,end=" ")

        else :
            print(0,end=" ")
    
    else :
        while target_num <= card and mid*2 < n :
            mid *= 2
            target_num = nlist[mid]

        if card in nlist[mid//2:mid+1]:
            print(1,end=" ")

        else :
            print(0,end=" ")
```

```python
# 10815 숫자카드 try 2

n = int(input())
nlist = sorted(list(map(int,input().split())))
m = int(input())
mlist = list(map(int,input().split()))

def cardcheck(target_num,start,end):
    mid = (start + end) // 2

    while mid <= end and mid >= start and start <= end :

        if nlist[mid] == target_num :
            return 1

        if nlist[mid] < target_num :
            start = mid

        if nlist[mid] > target_num :
            end = mid

        mid = (start + end) // 2

    return 0

for i in mlist:
    print(cardcheck(i,0,5),end=" ")
```

```python
# 3036 링

import math

n = int(input())

nlist = list(map(int, input().split()))

for idx in range(len(nlist)-1):

    gcd_num = math.gcd(nlist[0], nlist[idx+1])  ## 두수의 최대공약수 구함

    if gcd_num != 1:  ## 최대공약수가 1이아니면 최대공약수로 각 분자 분모값을 나눠주고 프린트
        first_num = nlist[0]//gcd_num  
        second_num = nlist[idx+1]//gcd_num
        
    else:  ## 최대공약수가 1이면 각 분자 분모값 그대로 프린트
        first_num = nlist[0]
        second_num = nlist[idx+1]

    print(f"{first_num}/{second_num}")
```

```python
# n 까지 파스칼 삼각형 만들기

cache = [[0 for _ in range(1001)] for _ in range(1001)]

for x in range(1001):
    for y in range(1001):
        if not cache[x][y]:
            if y == 0 or y == x :
                cache[x][y] = 1
            else :
                cache[x][y] = cache[x-1][y-1] + cache[x-1][y]



n, k = map(int,input().split())

print(cache[n][k]%10007)
```

```python
## 정수론 검문

from math import gcd
import sys

input = sys.stdin.readline

n = int(input())
number_list = []

for i in range(n):
    m = int(input())
    number_list.append(m)

number_list.sort()

diff_list = []
gcd_list = []
result = []

if n == 2 :
    target_number = number_list[1] - number_list[0]

    for i in range(1,int(target_number**0.5)+1):
        if target_number % i == 0:
            result.append(i)
            result.append(target_number//i)
    
else:
    for i in range(1,n):
        diff_list.append(number_list[i] - number_list[i-1])

    for i in range(1,n-1):
        gcd_list.append(gcd(diff_list[i-1],diff_list[i]))


    min_val = min(gcd_list)

    for i in range(1,int(min_val**0.5)+1):
        if min_val % i == 0 :
            result.append(i)
            result.append(min_val//i)

result = list(set(result))
result.sort()

for num in result[1:]:
    print(num,end=' ')

```
