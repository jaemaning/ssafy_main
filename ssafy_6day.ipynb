## 여러가지 내장함수

```python
def my_magic_func(n):
    return n*10

my_list = [1,2,3,4,5]

rlt = map(my_magic_func, my_list)

print(list(rlt))
```

```python
# zip 내장함수로 2개의 for 문을 동시에 돌릴 수 있다.

name_list = ["kim","park","jang","ryu","son"]
age_list = [25,22,33,15,24]

for name,age in zip(name_list,age_list):
    print(name,age)
```

```python
def recur():
    print("뿅")
    recur()

recur()
```

```python
def fac(n):
    if n == 1:
        return 1
    else:
        return n * fac(n - 1)

fac(5)
```

```python
# 패킹 *

x, y = 1, 2
z = 1, 2, 3

print(z,type(z))  # (1,2,3) tuple

a, *b = 1,2,3,4
print(a,b)


# 언패킹 *

def my_sum(a, b, c):
    return a + b + c

num_list = [10, 20, 30]

# rlt = my_sum(num_list[0], num_list[1], num_list[2])
rlt = my_sum(*num_list)

print(rlt)
print(*num_list)
```

```python
# 패킹 / 언패킹 *

def test(a,*args) :
    rlt = a
    for value in args:
        rlt += value

    return rlt


test(0)
test(1,2,3,4,5)

# 키워드 가변인자 **
   # *args : 포지션별 가변인자 , **kwargs : 키워드 가변인자
def test2(*args, **kwargs) : 
    print(*args, kwargs, type(kwargs))
    return kwargs

test2(1, 2, 3, 4, name='aiden', age=22)
```
